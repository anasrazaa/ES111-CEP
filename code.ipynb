```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing weights file
weights_data = pd.read_csv("weights_data.csv")
```

```python
# Taking random sample of size N=10 and applying formula for z-score
N = 10
z_scores = []
for i in range(1000):
    sample = np.random.choice(weights_data["weights"], size=N)
    z = (np.mean(sample) - np.mean(weights_data["weights"])) / (np.std(weights_data["weights"]) / np.sqrt(N))
    z_scores.append(z)

# Ploting the results
N, bins, patches = plt.hist(z_scores, bins=20, color= 'green', edgecolor = 'white')
colors = ['#FFF8DC', '#FFEBCD', '#FFE4C4', '#FFDEAD', '#F5DEB3', '#DEB887', '#D2B48C', '#BC8F8F', '#F4A460', '#DAA520', '#B8860B', '#CD853F', '#D2691E', '#A0522D', '#D2B48C', '#DEB887', '#F5DEB3', '#FFDEAD', '#F5DEB3', '#DEB887']
for i in range(len(N)):
    c = colors[i]
    patches[i].set_fc(c)
plt.title("Distribution of z-scores for N=10")
plt.xlabel("z-score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Taking random sample of size N=100 and applying formula for z-score
N = 100
z_scores = []
for i in range(1000):
    sample = np.random.choice(weights_data["weights"], size=N, replace=False)
    z = (np.mean(sample) - np.mean(weights_data["weights"])) / (np.std(weights_data["weights"]) / np.sqrt(N))
    z_scores.append(z)


# Ploting the results
N, bins, patches = plt.hist(z_scores, bins=20, edgecolor = 'white')
colors = ['#D8BFD8', '#DDA0DD', '#EE82EE', '#DA70D6', '#FF00FF', '#FF00FF', '#BA55D3', '#9370DB', '#8A2BE2', '#9400D3', '#9932CC', '#8B008B', '#800080', '#663399', '#4B0082', '#7B68EE', '#6A5ACD', '#483D8B', '#BA55D3', '#FF00FF']
for i in range(len(N)):
    c = colors[i]
    patches[i].set_fc(c)
plt.title("Distribution of z-scores for N=100")
plt.xlabel("z-score")
plt.ylabel("Frequency")
plt.show()
```
